categorical plots

```python
import seaborn as sns
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#bar plot n count plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x ='sex', hue = "smoker", data = tips)

# Show the plot
plt.show()
```

```python
#boxplot
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips)
```

```python
sns.boxplot(data=tips,orient='h')
```

```python
#violinplot
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
#strip plot
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
#swarmplot
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips)
```

```python
# Combining Categorical Plots
sns.violinplot(x="tip", y="day", data=tips)
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

```python
#catplot
sns.catplot(x="day", y="total_bill", kind="bar", data=tips)
```

```python
#relplot
sns.relplot(data=tips, x="total_bill", y="tip", hue="time",col="day") # scatter n lineplot  because of kind
```

```python
#Distribution plots
#displot
sns.displot(tips['total_bill'],bins=8,kde=True)
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(data=tips, x="total_bill")
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="layer")
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.pairplot(tips)
```

```python
#pairplot
sns.pairplot(tips,kind="hist",hue='sex')
```

```python
iris = sns.load_dataset("iris")
g = sns.pairplot(iris)
plt.show()
```

```python
#rugplot
sns.rugplot(tips['total_bill'])
```

```python
#Grid plot
#facetgrid
g = sns.FacetGrid(tips, col="time",  row="day",hue="smoker")
g = g.map(sns.regplot, "total_bill","tip").add_legend()
```

```python
#regression plots
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
# regression plots
sns.lmplot(x="total_bill", y="tip", data=tips,hue="sex",col='time',row='smoker')
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips,hue="sex",col='time')
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips,hue="sex")
```

```python

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection ='3d')

# defining axes
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)



ax.plot3D(x, y, z, 'green')   #lineplot
ax.set_title('3D line plots ')
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection ='3d')

# defining axes
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

c=x+y
ax.scatter(x,y,z,c=c)

ax.set_title('3D scatter plots ')
plt.show()
```
